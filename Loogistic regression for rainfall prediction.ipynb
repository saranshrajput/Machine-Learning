```python
import pandas as pd
import numpy as np
data=pd.read_csv('D:\Data_Sets\weather.csv')
data.info()
```

# Replacing nan values with mean of the column

```python
data.fillna(data.mean(), inplace=True)
data.info()
```

```python
import matplotlib.pyplot as plt
data.plot(x = 'MaxTemp' , y = 'MinTemp' , style = 'o')
plt.xlabel('MaxTemp')
plt.ylabel('MinTemp')
plt.figure(figsize=(20,20))
# plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
sns.heatmap(data.corr())
```

```python
data
```

```python
data.RainToday = [1 if each == 'Yes' else 0 for each in data.RainToday]
data.RainTomorrow = [1 if each == 'Yes' else 0 for each in data.RainTomorrow]
```

```python
data.head()
```

```python
x = data.drop([ 'WindGustDir', 'WindDir9am', 'WindDir3pm' , 'RainTomorrow'] , axis = 1)
y = data['RainTomorrow']
# y = y.values.reshape(-1, 1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression , LinearRegression
```

```python
x_train , x_test , y_train , y_test = train_test_split( x , y , test_size = 0.2 , random_state = 0) 
```

```python
classifier = LogisticRegression()
```

```python
classifier.fit(x_train , y_train)
```

```python
Predictions = classifier.predict(x_test)
```

```python
Predictions
```

```python
from sklearn.metrics import classification_report
classification_report(y_test,Predictions)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': Predictions})
df1 = df.head(25)
```

```python
df1
```

```python
df1.plot(kind='bar',figsize=(10,8))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, Predictions))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, Predictions))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, Predictions)))
```
